```python
import numpy as np
import pandas as pd
import MetaTrader5 as mt5
import plotly.express as px

from library.Constants import CURRENCY_PAIRS
import library.MetaTrader_Data as mtd
```

```python
PAIRS = CURRENCY_PAIRS
TIMEFRAME = mt5.TIMEFRAME_H1
BARS = 500
SMA_PERIOD = 50
```

```python
mtd.fill_data(PAIRS, TIMEFRAME, BARS)
```

```python

for symbol, df in mtd.data.items():
    df['sma'] = df['close'].rolling(SMA_PERIOD).mean()
    fig = px.line(df, x='time', y=['close', 'sma'], title=symbol, template="plotly_dark")
    fig.show()
```
