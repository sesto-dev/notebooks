```python
import numpy as np
import pandas as pd
import MetaTrader5 as mt5
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller

import matplotlib.pyplot as plt
import shared
```

```python
PAIRS = ["WTI", "BRN", 'USDJPY']
TIMEFRAME = mt5.TIMEFRAME_H1
BARS = 500
```

```python
shared.fill_data(PAIRS, TIMEFRAME, BARS)
print(shared.data)
```

```python
for symbol, data in shared.data.items():
    plt.figure(figsize=(10,6))
    plt.plot(data['close'])
    plt.title(f'{symbol} Close Price')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10,6))

# Loop through shared.data and plot each line on the same chart
for symbol, data in shared.data.items():
    # Normalize the close column
    normalized_close = data['close'] / data['close'].max()
    
    # Plot the normalized line
    ax.plot(normalized_close, label=symbol)

# Add title, x-axis label, y-axis label, and legend
ax.set_title('Close Prices')
ax.set_xlabel('Time')
ax.set_ylabel('Price')
ax.legend()

# Show the figure
plt.show()
```
