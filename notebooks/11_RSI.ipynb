```python
import numpy as np
import pandas as pd
import pandas_ta as ta
import MetaTrader5 as mt5
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller

import matplotlib.pyplot as plt
import shared
```

```python
PAIRS = ["WTI", "BRN", "USDJPY", "EURUSD", "GBPUSD"]
TIMEFRAME = mt5.TIMEFRAME_H1
BARS = 500
WINDOW_FACTOR = 14
```

```python
shared.fill_data(PAIRS, TIMEFRAME, BARS)
```

```python
for symbol, data in shared.data.items():
    rsi = ta.rsi(data['close'], length=WINDOW_FACTOR)
 
    plt.plot(rsi)
    plt.title(f'{symbol} Relative Strength Index')
    plt.xlabel('Time')
    plt.ylabel('Relative Strength Index')
    plt.show()
```
