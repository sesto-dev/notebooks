```python
import numpy as np
import pandas as pd
import MetaTrader5 as mt5
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller

import matplotlib.pyplot as plt
import shared
from library.Constants import CURRENCY_PAIRS
```

```python
PAIRS = CURRENCY_PAIRS
TIMEFRAME = mt5.TIMEFRAME_H1
BARS = 500
WINDOW_FACTOR = 50
```

```python
shared.fill_data(PAIRS, TIMEFRAME, BARS)
```

```python
for symbol, data in shared.data.items():
    rolling_kurt = data['close'].rolling(window=WINDOW_FACTOR).kurt()
    plt.plot(rolling_kurt)
    plt.title(f'{symbol} Rolling Kurtosis')
    plt.xlabel('Time')
    plt.ylabel('Rolling Kurtosis')
    plt.show()
```
