```python
import numpy as np
import pandas as pd
import MetaTrader5 as mt5
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller

import matplotlib.pyplot as plt
from backtester.data_handler import DataHandler
```

```python
SYMBOLS = ["WTI", "BRN"]
TIMEFRAME = mt5.TIMEFRAME_H1
BARS = 500
WINDOW_FACTOR = 50
```

```python
data_handler = DataHandler(SYMBOLS, TIMEFRAME, BARS)
data = data_handler.fetch_data()
```

```python
for symbol, symbol_data in data.items():
    rolling_kurt = symbol_data['close'].rolling(window=WINDOW_FACTOR).kurt()
    plt.plot(rolling_kurt)
    plt.title(f'{symbol} Rolling Kurtosis')
    plt.xlabel('Time')
    plt.ylabel('Rolling Kurtosis')
    plt.show()
```
