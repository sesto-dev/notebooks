```python
import sys
sys.path.append("..")

import numpy as np
import pandas as pd
import MetaTrader5 as mt5

from sesto.constants import CURRENCY_PAIRS
import sesto.metatrader.data as mtd
from sesto.plot import plot_plotly
```

```python
PAIRS = CURRENCY_PAIRS
TIMEFRAME = mt5.TIMEFRAME_H1
BARS = 300
SMA_PERIOD = 21
NUM_STD = 2
```

```python
mtd.fill_data_pos(PAIRS, TIMEFRAME, BARS)
```

```python
def calculate_bollinger_bands(data, window=SMA_PERIOD, num_std_dev=NUM_STD):
    rolling_mean = data['close'].rolling(window=window).mean()
    rolling_std = data['close'].rolling(window=window).std()
    
    data['BBM'] = rolling_mean
    data['BBU'] = rolling_mean + (rolling_std * num_std_dev)
    data['BBL'] = rolling_mean - (rolling_std * num_std_dev)
```

```python
# Loop over each symbol in the shared.data dictionary
for symbol, df in mtd.data.items():
    # Calculate Bollinger Bands
    calculate_bollinger_bands(df)
```

```python
for symbol, df in mtd.data.items():
    plot_plotly(df, symbol, ['close', 'BBM', 'BBU', 'BBL'])
```
