```python
import numpy as np
import pandas as pd
import pandas_ta as ta
import MetaTrader5 as mt5
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller

import matplotlib.pyplot as plt
from backtester.data_handler import DataHandler
```

```python
SYMBOLS = ["WTI", "BRN", "USDJPY", "EURUSD", "GBPUSD"]
TIMEFRAME = mt5.TIMEFRAME_H1
BARS = 500
WINDOW_FACTOR = 14
```

```python
data_handler = DataHandler(SYMBOLS, TIMEFRAME, BARS)
data = data_handler.fetch_data()
```

```python
for symbol, symbol_data in data.items():
    rsi = ta.rsi(symbol_data['close'], length=WINDOW_FACTOR)
 
    plt.plot(rsi)
    plt.title(f'{symbol} Relative Strength Index')
    plt.xlabel('Time')
    plt.ylabel('Relative Strength Index')
    plt.show()
```
